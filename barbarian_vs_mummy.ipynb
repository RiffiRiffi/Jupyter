```python
import random
class Personaje():
    __nombre=""
    __vida=0
    __ataque=0
    __defensa=0
    def __init__(self, nombre, vida, ataque, defensa):
        self.__nombre= nombre
        self.__vida= vida
        self.__ataque=ataque
        self.__defensa=defensa
    
    @property
    def nombre(self):
        return self.__nombre
    @nombre.setter
    def nombre(self, newName):
        self.__nombre = newName
        
    @property
    def vida(self):
        return self.__vida
    @vida.setter
    def vida(self, newLife):
        self.__vida = newLife
        
    @property
    def ataque(self):
        return self.__ataque
    @ataque.setter
    def ataque(self, newAtack):
        self.__ataque = newAtack
    
    @property
    def defensa(self):
        return self.__defensa
    @defensa.setter
    def defensa(self, newDefense):
        self.__defensa = newDefense
        
    def __str__(self):
        return "Hola"
    
    def estarVivo(self):
        self.vivir = True
        if self.__vida <= 0:
            self.vivir = False
            return self.vivir
        else:
            pass
        
        
    
    def atacar(self):
        numImpactos=0
        for i in range(self.__ataque):
            i = Dado()
            if i.tira()>3:
                numImpactos+=1
        return numImpactos
            
            

class Momia(Personaje):
    def __init__(self, nombre, vida, ataque, defensa):
        super().__init__(nombre, vida, ataque, defensa)
    def __str__(self):
        return "Se ha creado la Momia {}".format(self.nombre)
    def defender(self, numImpactos):
        recibidos = numImpactos
        defendidos=0
        for i in range(self.defensa):
            i = Dado()
            if i.tira() == 6:
                defendidos+=1
        total = recibidos - defendidos
        if total < 0 or total == 0:
            total = 0
            print(">>La momia {} pudo bloquear totalmente el ataque y queda con {} de vida".format(self.nombre, self.vida))
        else:
            if (self.vida - total)<0:
                print(">>La momia {} no pudo bloquear {} impactos y queda con 0 de vida".format(self.nombre, total))
            else:
                self.vida = self.vida - total
                print(">>La momia {} no pudo bloquear {} impactos y queda con {} de vida".format(self.nombre, total, self.vida))
            
        
        

class Barbaro(Personaje):
    def __init__(self, nombre, vida, ataque, defensa):
        super().__init__(nombre, vida, ataque, defensa)
    def __str__(self):
        return "Se ha creado el Barbaro {}".format(self.nombre)
    def defender(self, numImpactos):
        recibidos = numImpactos
        defendidos=0
        for i in range(self.defensa):
            i = Dado()
            if i.tira() == 6:
                defendidos+=1
        total = recibidos - defendidos
        if total <  0 or total ==0:
            total = 0
            print(">>El barbaro {} pudo bloquear totalmente el ataque y queda con {} de vida".format(self.nombre, self.vida))
        else:
            if (self.vida - total)<0:
                print(">>El barbaro {} no pudo bloquear {} impactos y queda con 0 de vida".format(self.nombre, total))
            else:
                self.vida = self.vida - total
                print(">>El barbaro {} no pudo bloquear {} impactos y queda con {} de vida".format(self.nombre, total,
                                                                                                   self.vida))
        

        
class Dado():
    def tira(self):
        resultadoDado= random.randint(1, 6)
        return resultadoDado
    
barb = Barbaro(input("Cual es tu nombre, guerrero? "), 5, 3, 3)
momi = Momia ("Tutankamon", 6, 2, 2)

turnos = int(input("Cuantos turnos quieres jugar? "))
while turnos > 0 :
    momi.defender(barb.atacar())
    if momi.estarVivo() == False:
        break
    barb.defender(momi.atacar())
    if barb.estarVivo() == False:
        break
    turnos-=1



if turnos == 0 :
    print("Los luchadores han quedado empatados!!!")
else:
    if barb.vida < momi.vida:
        print (">GANADORA MOMIA {}\n   >Momia {} con {} de vida\n   >Barbaro {} con {} de vida".format(momi.nombre,
                                                                                                       momi.nombre,
                                                                                                       momi.vida,
                                                                                                       barb.nombre,
                                                                                                       barb.vida))
    else:
        print (">GANADOR BARBARO {}\n   >Barbaro {} con {} de vida\n   >Momia {} con {} de vida".format(barb.nombre,
                                                                                                        barb.nombre,
                                                                                                        barb.vida,
                                                                                                        momi.nombre,
                                                                                                        momi.vida))
```
